### KNN tanpa librari

```python
import numpy as np
def knn1(arr, x, wisata, k=1):
  knn = []
  for i in arr:
    dist = np.array([np.linalg.norm(x[j] - i) for j in range(len(x))])
    idx_k3n = np.where(dist <= dist[np.argsort(dist)])
    k3n = []
    for j in idx_k3n[0]:
      if len(k3n) < k:
        k3n.append(wisata[j])
    knn.append(k3n)
  return knn
```
